```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime


with open('./log_kafka.csv','r') as f:
    line = f.readline()
    timestamp = []
    cpu = []
    mem = []
    
    while line:
        words = line.split(",")
        timestamp.append(datetime.strptime(words[0],'%H:%M:%S.%f').time())
        cpu.append(float(words[1]))
        mem.append(float(words[2]))
        line = f.readline()
    kafkaDF = pd.DataFrame({'time':timestamp,'cpu':cpu, 'mem':mem})
kafkaDF.plot()
```

```python
a = 2. # shape
n = 1.
s = np.random.weibull(a, 1000)
import matplotlib.pyplot as plt
x = np.arange(1,100.)/50.

def weib(x,n,a):
    return (a / n) * (x / n)**(a - 1) * np.exp(-(x / n)**a)

count, bins, ignored = plt.hist(np.random.weibull(a,1000))

scale = count.max()/weib(x, n, a).max()
plt.plot(x, weib(x, n, a)*scale)
plt.show()
```
